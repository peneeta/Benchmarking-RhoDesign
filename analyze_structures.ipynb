## Compute RMSD and TM for all generated structures (Temp = 1)

```python
temp1_2au4 = "rhofold_pred/temp1/2AU4"
rmsd_2au4, tm_2au4 = CalculateRMSD_TM_Multiple(ref_2au4, temp1_2au4)
```

```python
temp1_5kpy = "rhofold_pred/temp1/5KPY"
rmsd_5kpy, tm_5kpy = CalculateRMSD_TM_Multiple(ref_5kpy, temp1_5kpy)
```

```python
temp1_6e8t = "rhofold_pred/temp1/6E8T"
rmsd_6e8t, tm_6e8t = CalculateRMSD_TM_Multiple(ref_6e8t, temp1_6e8t)
```

```python
xlabels = ["2AU4", "5KPY", "6E8T"]
```

```python
data = [rmsd_2au4, rmsd_5kpy, rmsd_6e8t]
fig, ax = plt.subplots(figsize=(7, 6))

parts = ax.violinplot(data, positions=np.arange(len(xlabels)), 
                      showmeans=True, showmedians=True)

# Customize violin colors
color = "#6B39FF"
for i, pc in enumerate(parts['bodies']):
    pc.set_facecolor(color)
    pc.set_alpha(0.7)

# Change mean and median marker colors to black
parts['cmeans'].set_color('black')
parts['cmeans'].set_linewidth(1)
parts['cmedians'].set_color('black')
parts['cmedians'].set_linewidth(1)

parts['cbars'].set_color('black')
parts['cbars'].set_linewidth(0.5)
parts['cmaxes'].set_color('black')
parts['cmaxes'].set_linewidth(0.5)
parts['cmins'].set_color('black')
parts['cmins'].set_linewidth(0.5)


# Display mean values on the plot
for i, arr in enumerate(data):
    mean_val = np.mean(arr)
    ax.text(i, mean_val, f'{mean_val:.4f}', 
            ha='center', va='bottom', fontsize=10)

# Customize plot
ax.set_xlabel('Aptamers', fontsize=12)
ax.set_ylabel('RMSD (Å)', fontsize=12)
ax.set_title('Mean RMSD Scores for Generated Aptamers', fontsize=14)
ax.set_xticks(np.arange(len(xlabels)))
ax.set_xticklabels(xlabels)
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('./figs/rmsd_generated.png', dpi=300, bbox_inches='tight')
plt.show()
```
